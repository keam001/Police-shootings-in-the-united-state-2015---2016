# The subplot() Function

- The layout is organized in rows and columns, which are represented by the first and second argument.

- The third argument represents the index of the current plot.

# BARCHAT IN PANDAS
- A bar chart (or bar graph) is a type of data visualization that uses rectangular bars to represent data values.
- Each bar's length or height is proportional to the value it represents, 
- allowing for easy comparison between categories.

# Key Features of a Bar Chart:
- Bars: Represent individual categories (e.g., countries) and their values (e.g., population).

- Axes:
The horizontal axis (x-axis) typically represents categories (e.g., country names).

- The vertical axis (y-axis) represents numerical values (e.g., population size).

- Bar Height/Length: Corresponds to the value of the category (the taller or longer the bar, the higher the value).

```python
# cREATING bARs
import matplotlib.pyplot as plt
# Sample data: Country names and their populations
countries  =['china','india','USA', 'indonesia','pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]
plt.figure(figsize=(10,6))

#Adding title and labels

bars = plt.bar(countries,population, color='black')
plt.title('Population of countries', size=16)
plt.xlabel ('COUNTRY', size=12)
plt.ylabel ('POUPLATION', size=12)
# Adding labels at the top of each bar
for bar in bars:
    height = bar.get_height()  # Get the height of the bar
    # Add text annotation on the bar
    plt.text(bar.get_x() + bar.get_width() / 2.,  # x position of text
             height * 1.01,  # y position of text, slightly above the bar
             f'{height:,} ',  # formatted text to display
             ha='center',  # horizontal alignment
             va='bottom',  # vertical alignment
             fontsize=10)  # font size

```

```python
 # cREATING bARs
import matplotlib.pyplot as plt
# Sample data: Country names and their populations
countries  =['china','india','USA', 'indonesia','pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]
plt.figure(figsize=(10,6))
bars = plt.barh(countries,population, color='gold')
plt.title('Population of countries', size=16)
plt.ylabel ('COUNTRY', size=12)
plt.xlabel ('POUPLATION', size=12)

for bar in bars:
    width = bar.get_width()  # Get the width of the bar (length in horizontal bar)
    height = bar.get_height()  # Get the height of the bar (thickness of the bar)
    
    # Add text annotation on the bar
    plt.text(width * 1.01,  # x position of text, slightly to the right of the bar
             bar.get_y() + height / 2.,  # y position of text, center of the bar
             f'{width:,} ',  # formatted text to display
             ha='left',  # horizontal alignment to the left of the text
             va='center',  # vertical alignment to the center of the bar
             fontsize=10)  # font size
```

## Race and Gender Analysis

```python
race_gender_count = pshoot.groupby(['race', 'gender']).size().reset_index(name='Count')
print(race_gender_count)
```

```python
pivot_race_gender = pshoot.pivot_table(index='race', columns='gender', aggfunc='size', fill_value=0)
print(pivot_race_gender)
```

```python
import matplotlib.pyplot as plt

# Plot a grouped bar chart
pivot_race_gender.plot(kind='bar', stacked=True)
plt.title('Race and Gender Distribution')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
```

```python
# Breakdown by race and gender (mr Dan answer)
race_gender_breakdown = pshoot.groupby(['race', 'gender']).size()

race_gender_breakdown
```

```python
location_count = pshoot['state'].value_counts().head(5)
print(location_count)
```

```python
# Count incidents by state (Mr Dan solution)
state_counts = pshoot['state'].value_counts()

state_counts
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart for the top 5 locations
location_count.plot(kind='bar', color='skyblue')
plt.title('Top 5 Locations by Count')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Filter the DataFrame for the top 5 locations
top_5_data = pshoot[pshoot['state'].isin(location_count.index)]

# Group by location, race, and gender
race_gender_top_5 = top_5_data.groupby(['state', 'race', 'gender']).size().unstack()

# Plot the race and gender distribution for the top 5 locations
race_gender_top_5.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Race and Gender Distribution in Top 5 Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Time-Based Analysis
-- In which months do most police shootings occur?

```python
# Count incidents by month
#incidents_per_month = pshoot['date'].dt.month.value_counts()

#incidents_per_month

pshoot['date'] = pd.to_datetime(pshoot['date'], errors='coerce')  # Convert to datetime, handling errors
```

```python
incidents_per_month = pshoot['date'].dt.month.value_counts()

incidents_per_month
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart
incidents_per_month.plot(kind='bar', color='coral')
plt.title('Incidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot a bar chart for incidents per year
incidents_per_year.plot(kind='bar', color='lightgreen')
plt.title('Number of Incidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```
